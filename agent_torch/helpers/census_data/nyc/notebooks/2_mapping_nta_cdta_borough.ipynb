## Relationship Mapping

1. NTA_ID: [CountyFIPS, CountyName, NTAType]
2. CountyName: [NTA_List]

```python
map_df = df.filter(items=['NTACode', 'CountyFIPS', 'BoroName', 'NTAType'])

dct = map_df.to_dict()
```

```python
nta_metadata = dict(zip(map_df.NTACode, list(zip(map_df.CountyFIPS, map_df.BoroName, map_df.NTAType))))

np.save('nta_metadata.npy', nta_metadata)
```

```python
county_nta = {}
for county in map_df.CountyFIPS.unique():
    nta_list = list(map_df[map_df['CountyFIPS'] == county]['NTACode'].unique())
    print("County: ", county ," Num ntas: ", len(nta_list))

    county_nta[county] = nta_list

np.save('county_to_nta_map.npy', county_nta)
```

```python
len(map_df['NTACode'].unique())
county = 5
filtered_df = map_df[map_df['CountyFIPS'] == county]['NTACode'].unique()

len(filtered_df)

nta_set = set()
for cty in county_nta:
    nta_set = nta_set.union(set(county_nta[cty]))

print(len(nta_set))
```
