```python
age_list = ['U19','20t29', '30t39', '40t49', '50t64', '65A']
def map_age_to_range(age):
    if age < 20:
        return 'U19'
    elif 20 <= age <= 29:
        return '20t29'
    elif 30 <= age <= 39:
        return '30t39'
    elif 40 <= age <= 49:
        return '40t49'
    elif 50 <= age <= 64:
        return '50t64'
    else:
        return '65A'


```

```python
df_gender_combined = pd.DataFrame()
```

```python
df_gender_combined['U19'] = df_gender.loc[:, 0:19].sum(axis=1)
df_gender_combined['20t29'] = df_gender.loc[:, 20:29].sum(axis=1)
df_gender_combined['30t39'] = df_gender.loc[:, 30:39].sum(axis=1)
df_gender_combined['40t49'] = df_gender.loc[:, 40:49].sum(axis=1)
df_gender_combined['50t64'] = df_gender.loc[:, 50:64].sum(axis=1)
df_gender_combined['65A'] = df_gender.loc[:, 65:100].sum(axis=1)
```

```python
df_gender_combined['area'] = df_gender['area']
df_gender_combined['gender'] = df_gender['gender']
```

```python
df_gender_combined
```

This is the resulting dataframe we use as input for ageand gender

```python
df_gender_combined = df_gender_combined.melt(
        id_vars=["area", "gender"], var_name="age", value_name="count"
    )
df_gender_combined
```
